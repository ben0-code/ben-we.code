```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime
from datetime import timedelta
import pandas as pd
# import pickle
# import time

def get_ps5_prices():

    url = "https://ci.coinafrique.com/search?category=&keyword=ps5"

    headers = {
        "User-Agent": (
            "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
            "AppleWebKit/537.36 (KHTML, like Gecko) "
            "Chrome/79.0.3945.74 Safari/537.36 "
            "Edg/79.0.309.43"
        )
    }

    Article = []
    prix = []
    localisaton = []
    tpls = []



    response = requests.get(url, headers=headers)

    temps = datetime.now()

    jeux = response.text
    soup = BeautifulSoup(jeux, 'html.parser')
    soup.find_all("div", class_="col s6 m4 l3")
    ben = soup.find_all("div", class_="col s6 m4 l3")
    for fruit in ben:
        element_trouve = fruit.select_one(".ad__card-description")
        Article.append(element_trouve.get_text())
        element_ben = fruit.select_one(".ad__card-price")
        prix.append(element_ben.get_text())
        element_gta = fruit.select_one(".ad__card-location")
        localisaton.append(element_gta.get_text())
        element_time = fruit.select_one(".time")
        pub = fruit.select_one("span:not([class])")
        exo = pub.get_text()
        tps_r = temps
        if exo == "jours":
            tps_r = temps - timedelta(days=int(int(element_time.get_text()))) 
        elif exo == "mois":
            tps_r = temps - timedelta(days=int(int(element_time.get_text())*30))
        elif exo == "minute":
            tps_r = temps - timedelta(minutes=int(int(element_time.get_text())))
        elif exo == "heures":
           tps_r = temps - timedelta(hours=int(int(element_time.get_text())))
        tpls.append(tps_r.isoformat())

    
    data = {
        'Article': Article,
        'prix': prix,
        'localisation': localisaton,
        'tpls': tpls,

    }
        
    df = pd.DataFrame(data)

    # print(df)
    
    return df
pur = get_ps5_prices()
pur
```

```python
import pickle
import time
import pandas as pd

last_df = None  

while True:
    new_df = get_ps5_prices()  
    if last_df is None or not new_df.reset_index(drop=True).equals(last_df.reset_index(drop=True)):
        with open("ps5-dataframe.pickle", "wb") as f:
            pickle.dump(new_df, f)

        print("tableau mis à jour :")
        print(new_df.to_string(index=False))

        last_df = new_df.copy()
    
    print("attente de 5 minutes\n")
    time.sleep(300)
```
